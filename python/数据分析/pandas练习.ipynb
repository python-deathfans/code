## 4、apply函数

```python
data_path = './exercise_data/US_Crime_Rates_1960_2014.csv'
```

```python
crime = pd.read_csv(data_path)
crime.head()
```

```python
crime.info()
```

由于year的类型是int64,可以把时间转换成datetime64类型

```python
crime.Year = pd.to_datetime(crime.Year, format='%Y')
crime.info()
```

```python
crime.describe()
```

将Year设置为索引

```python
crime = crime.set_index('Year', drop=True)
crime
```
